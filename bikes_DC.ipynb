```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
url = "/Users/matthewstevens/Desktop/day.csv"
days = pd.read_csv(url, parse_dates=[1])
```

```python
days.head()
```

```python
del days["instant"]
days["dteday"] = pd.to_datetime(days["dteday"])
days = days.set_index("dteday")
days.head()
```

```python
def fix_season(x):
    if x == 1:
        return 4
    else:
        return x - 1

days["season"] = days.season.apply(fix_season)
```

```python
days.tail(50)
```

```python
days.dtypes
```

```python
days.groupby(days.index.month).cnt.sum().sort_values(ascending=False)
```

```python
weekday_cnt = days.loc[:, ["weekday", "cnt"]]
```

```python
weekday_cnt.head()
```

```python
def get_day(x):
    d = {
        0: "Sunday",
        1: "Monday",
        2: "Tuesday",
        3: "Wednesday",
        4: "Thursday",
        5: "Friday",
        6: "Saturday",
    }
    
    return d.get(x)

weekday_cnt["weekday"] = weekday_cnt["weekday"].apply(get_day)
```

```python
weekday_cnt.head()
```

```python
plt.title("Most Popular Days to Rent Bikes in D.C.")
plt.ylabel("Number of Bike Rentals")
weekday_cnt.groupby(["weekday"]).cnt.sum().sort_values(ascending=False).plot(kind="bar")
```

```python
seasons = days[["season", "cnt"]]
seasons.head()
```

```python
def fix_season(x):
    s = ["Spring", "Summer", "Fall", "Winter"]
    return s[x-1]

seasons["season"] = seasons.season.apply(fix_season)
```

```python
seasons.head()
```

```python
seasons.groupby(["season"]).cnt.sum().sort_values().plot(kind="bar")
plt.title("Most Popular Seasons for Bike Rentals\nIn D.C.")
plt.ylabel("Seasons")
plt.xlabel("Number of Rentals")
plt.show()
```

```python
h5 = days[["weathersit", "cnt"]]
h5.head(3)
```

```python
def get_condition(w):
    if w == 1:
        return "clear"
    if w == 2:
        return "mist/clouds"
    if w == 3:
        return "light rain/light snow"
    if w == 4:
        return "heavy rain/heavy snow"
    
h5["weathersit"] = h5["weathersit"].apply(get_condition)
h5.head()
```

```python
h5.groupby("weathersit").cnt.sum().sort_values().plot.bar()
```

```python
h6 = days[["workingday", "cnt"]]
def fix_day(x):
    if x == 0:
        return "Day off"
    else:
        return "Working Day"
h6["workingday"] = h6["workingday"].apply(fix_day)
h6 = h6.groupby("workingday").cnt.sum()
h6.plot.bar()
plt.title("Bike Rentals")
plt.xlabel("Kind of Day")
plt.ylabel("Number of Bike Rentals")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor as DTR
bike_model = DTR(random_state=1)
y = days.cnt
desired = days.columns[(days.columns != "yr") & (days.columns != "casual") & (days.columns != "registered") & (days.columns != "cnt")]
```

```python
X = days[desired]
X.head()
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
bike_model.fit(train_X, train_y)
```

```python
val_predictions = bike_model.predict(val_X)
```

```python
print(val_predictions)
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE: ", mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.ensemble import RandomForestRegressor as RFG
bike_forest_model = RFG(random_state=1)
bike_forest_model.fit(train_X, train_y)
forest_predictions = bike_forest_model.predict(val_X)
print("Forest MAE: ", mean_absolute_error(val_y, forest_predictions))
```
